```python
import pandas as pd

# Load the CSV file from the specified path
df = pd.read_csv('/content/motor temp 5.CSV')

# Filter rows with 'Rx' in the 'Direction' column
rx_rows = df[df['Direction'] == 'Rx']

# Select only the required columns
output = rx_rows[['Direction', 'Time', 'Data(HEX)']].copy()

# Function to check the 4th byte and, if it matches any specified value, extract and rearrange the 5th and 6th bytes
def check_and_rearrange(data_hex):
    try:
        bytes_list = data_hex.split()
        if len(bytes_list) < 6:
            return None, None, None, None, None, None, None  # If there are not enough bytes, return None for all
        byte_4 = bytes_list[3]
        byte_5 = bytes_list[4]
        byte_6 = bytes_list[5]
        rearranged_bytes = byte_6 + byte_5

        if byte_4 == '0d':
            return rearranged_bytes, int(rearranged_bytes, 16) / 10, None, None, None, None, None
        elif byte_4 == '04':
            return None, None, int(rearranged_bytes, 16), None, None, None, None
        elif byte_4 == '05':
            return None, None, None, int(rearranged_bytes, 16), None, None, None
        elif byte_4 == '11':
            return None, None, None, None, int(rearranged_bytes, 16), None, None
        elif byte_4 == '06':
            return None, None, None, None, None, int(rearranged_bytes, 16), None
        elif byte_4 == '03':
            return None, None, None, None, None, None, int(rearranged_bytes, 16)/10
        else:
            return None, None, None, None, None, None, None
    except:
        return None, None, None, None, None, None, None

# Apply the function to the 'Data(HEX)' column
output['RearrangedBytes'], output['DecimalValue'], output['InputVoltage'], output['OutputVoltage'], output['AnalogA1'], output['MechanicalSpeed'], output['AnalogOutputCurrent'] = zip(*output['Data(HEX)'].apply(check_and_rearrange))

# Merge the final output back into the original DataFrame
df = df.merge(output[['Time', 'RearrangedBytes', 'DecimalValue', 'InputVoltage', 'OutputVoltage', 'AnalogA1', 'MechanicalSpeed', 'AnalogOutputCurrent']], on='Time', how='left')

# Save the updated DataFrame back to a new CSV file
df.to_csv('/content/temp 5_3.CSV', index=False)

# Display the updated DataFrame
df
```
