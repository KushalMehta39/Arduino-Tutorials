```python
import os
import pandas as pd
```

```python
all_data = pd.read_csv("No_Delay_amp_topic.csv")
all_data['.data'] = all_data['.data'].apply(lambda x: round(x, 1))
# all_data.to_csv('filename_rounded.csv')
all_data.head()
```

```python
def get_timem(address):
    return address.replace(":", ".")
def get_timeM(address):
    return address.split(".")[0] + address.split(".")[1] + address.split(".")[2]
def get_timeMM(address):
    return address.split("/")[3].strip(" ")
all_data['time_m'] = all_data['time'].apply(lambda x: f"{get_timem(x)}")
all_data['time_M'] = all_data['time_m'].apply(lambda x: f"{get_timeM(x)}")
all_data['time_M'] = all_data['time_M'].apply(lambda x: f"{get_timeMM(x)}")
all_data.head()
```

```python
all_data = all_data.drop('time', axis = 1)
all_data = all_data.drop('time_m', axis = 1)
all_data.head()
```

```python
size = list(all_data['time_M'])
dataa = list(all_data['.data'])
print(len(size))
print(size[920])
print(dataa[920])
print(size)
```

```python
for a, i in enumerate(size):
    size[a] = int(size[a])
    dataa[a] = float(dataa[a])
print(size)
indices = [i for i, x in enumerate(size) if x == 2621]
print(dataa)
print(indices)
for i in indices:
    print(dataa[i])
```

```python
# mc_stan = {}
from collections import defaultdict
# mc_stan = {key: [] for key in range(size[0], size[-1])}
# for i,a in zip(size, dataa):
mc_stan = defaultdict(list)

# Use map to iterate over elements of 'a' and 'b' simultaneously and update the result_dict
for key, value in zip(size, dataa):
    if value in dataa:
        mc_stan[key].append(value)

# Convert the defaultdict to a regular dictionary
ans = dict(mc_stan)
print(ans)
```

```python
import random
for i in ans:
    if len((ans[i]))==4:
        continue
    elif len(ans[i])<4:
        while len(ans[i])<4:
            ans[i].append(ans[i]
                          [len(ans[i])-1])
        # continue
    elif len((ans[i]))>4:
        while len(ans[i])!= 4:
            rndm_element = random.choice(ans[i])
            ans[i].remove(rndm_element)

# for key in ans:
#     sumlya = sum(ans[key])
#     sumlya = sumlya/len(ans[key])
#     ans[key] = sumlya
superans = {}
for a in ans:
    one = a*100
    two = one+25
    three = one+50
    four = one + 75
    superans[one] = ans[a][0]
    superans[two] = ans[a][1]
    superans[three] = ans[a][2]
    superans[four] = ans[a][3]
print(superans)
```

```python
data_list = [(time, gearValue) for time, gearValue in superans.items()]
new_df = pd.DataFrame(data_list, columns=['time', 'data'])
new_df.head()
```

```python
def get_timeINFotmat(intTime):
    time_str = str(intTime)
    hours = time_str[:2]
    minutes = time_str[2:4]
    seconds = time_str[4:6]
    microseconds = time_str[6:]
    return f"{hours}:{minutes}:{seconds}:{microseconds}"

new_df['time'] = new_df['time'].apply(lambda x: f"{get_timeINFotmat(x)}")
new_df.head()
# new_df.to_csv('file.csv', index = False)
```
